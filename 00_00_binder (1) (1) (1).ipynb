- 'Duos matches' = variable numerica que identifica el numero de partidas jugadas en modo de juego duos
- 'Duos kills' = variable numerica que identifica el numero de eliminaciones realizadas por jugador en todas las partidas que ha jugado en el modo de juego Duos

- 'Squads top1' = variable numerica que identifica el numero de ganadas en el modo de juego Squads

# Pregunta 1
### Hay una relacion entre minutos jugados y victorias en el modo de juego solo?

```python
sns.lmplot(x = 'Solo minutesPlayed',
           y = 'Solo top1',
           data = df,
           fit_reg=True)
```

# Pregunta 2
### Tiene que ver el numero de partidas jugadas en squads con el porcentaje de victorias?

```python
sns.lmplot(x = 'Squads matches',
           y = 'Squads winRatio',
           data = df,
           fit_reg=False)
```

# Pregunta 3
## Que cantidad de victorias es lo mas comun que tengan los jugadores de pc

```python
from matplotlib.pyplot import figure
figure(figsize=(30,8),dpi=80)
plt.subplot(1,3,1)
plt.hist('Solo top1', bins=9,data = df, color = 'r')

plt.xlabel('Solo Wins')
plt.ylabel('Frecuencia')
plt.title('Solo')
plt.subplot(1,3,2)
plt.hist('Duos top1', bins=7,data = df,color= 'm')

plt.xlabel('Duos Wins')
plt.ylabel('Frecuencia')
plt.title('Duos')
plt.subplot(1,3,3)
plt.hist('Squads top1', bins=10,data = df, color= 'k')

plt.xlabel('Squads Wins')
plt.ylabel('Frecuencia')
plt.title('Squads')
plt.show()
```
